```python
from pydub import AudioSegment
song = AudioSegment.from_wav("data/lagu_sedih.wav")
start_minute = 0
start_second = 23
end_minute = 1
end_second = 23
start_time = start_minute * 60 * 1000 + start_second * 1000
end_time = end_minute * 60 * 1000 + end_second * 1000
extract = song[start_time:end_time]
extract.export("data/slow_song.wav", format="wav")
print("Selesai memotong suara")
```
